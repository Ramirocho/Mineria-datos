```python
# Ahora me interesa tener una columna agrupando los wilderness area en una sola  para esto tendremos
# un arreglo con las columnas con estas sumaremos ya que tienen un numero y en las otras 0 asi que podre obtener ese indice
Columnas_wilderness = ['Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3','Wilderness_Area4']
data["Wilderness_Area"] = data[Columnas_wilderness].apply(sum, axis=1) 
data.Wilderness_Area.describe()
```

```python
#Aca usamos idxmax que trae el id d ela columna y aplico una lambda que en python es una variable que no defino
# ya que tengo distintos tipos de argumentos como int y caracter  esta se usa como funcion anonima
#hago un split para recortar el wilderness area ya que sino lo hiciera me saldira wildernessarea1,2,3 dependiendo y solo
#quiero el indice por eso el menos 1 igual si pusiera menos dos me traeria la letra a eso depende de la posicion
data["Wilderness_Area"] = data[Columnas_wilderness].idxmax(axis=1)
data["Wilderness_Area"] = data["Wilderness_Area"].apply(lambda x: x.split("Wilderness_Area")[-1])
data.Wilderness_Area.head()
```

```python
#ahora crearemos la rutina de limpieza con los comandos que ya ejecutamos anteriormente
def limpieza_dataset(dataset):
    Columnas_soil = ['Soil_Type1', 'Soil_Type2', 'Soil_Type3',
       'Soil_Type4', 'Soil_Type5', 'Soil_Type6', 'Soil_Type7', 'Soil_Type8',
       'Soil_Type9', 'Soil_Type10', 'Soil_Type11', 'Soil_Type12',
       'Soil_Type13', 'Soil_Type14', 'Soil_Type15', 'Soil_Type16',
       'Soil_Type17', 'Soil_Type18', 'Soil_Type19', 'Soil_Type20',
       'Soil_Type21', 'Soil_Type22', 'Soil_Type23', 'Soil_Type24',
       'Soil_Type25', 'Soil_Type26', 'Soil_Type27', 'Soil_Type28',
       'Soil_Type29', 'Soil_Type30', 'Soil_Type31', 'Soil_Type32',
       'Soil_Type33', 'Soil_Type34', 'Soil_Type35', 'Soil_Type36',
       'Soil_Type37', 'Soil_Type38', 'Soil_Type39', 'Soil_Type40']
    Columnas_wilderness = ['Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3','Wilderness_Area4']
    dataset["Tipo_Soil"] = dataset[Columnas_soil].idxmax(axis=1)
    dataset["Tipo_Soil"] = dataset["Tipo_Soil"].apply(lambda x: x.split("Soil_Type")[-1])
    dataset["Wilderness_Area"] = dataset[Columnas_wilderness].idxmax(axis=1)
    dataset["Wilderness_Area"] = dataset["Wilderness_Area"].apply(lambda x: x.split("Wilderness_Area")[-1])
    dataset = dataset.drop(Columnas_wilderness, axis=1)#borramos como ya tenia capturado los nombres en un arreglo lo llame
    dataset = dataset.drop(Columnas_soil, axis=1)#llame los nombres del arreglo ya creado
    return dataset
```

```python
data=limpieza_dataset(data)# ejecutamos la rutina
```
